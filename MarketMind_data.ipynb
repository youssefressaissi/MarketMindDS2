**Extracting likes and comments from the engagement column**

```python
import re

# Function to extract likes and comments
def extract_engagement(dataframe):
    dataframe['Likes'] = dataframe['Engagement'].str.extract(r'(\d+) Likes').astype(float)
    dataframe['Comments'] = dataframe['Engagement'].str.extract(r'(\d+) Comments').astype(float)
    dataframe.drop(columns=['Engagement'], inplace=True)  # Remove the old column

# Apply to both datasets
extract_engagement(tunisia_df)
extract_engagement(global_df)

# Verify results
print("\n Tunisia Dataset - Engagement Extracted!")
print(tunisia_df[['Likes', 'Comments']].head())

print("\n Global Dataset - Engagement Extracted!")
print(global_df[['Likes', 'Comments']].head())
```

**Visualizing Rating Distribution + Normality Test**

**Using histograms & KDE plots and running Shapiro-Wilk normality tests**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import shapiro

# Function to visualize distribution & test normality
def visualize_rating_distribution(dataframe, column_name, dataset_name):
    plt.figure(figsize=(12, 5))

    # Histogram + KDE
    sns.histplot(dataframe[column_name].dropna(), kde=True, bins=30, color="royalblue", edgecolor="black")
    plt.axvline(dataframe[column_name].mean(), color='red', linestyle='dashed', label=f'Mean: {dataframe[column_name].mean():.2f}')
    plt.axvline(dataframe[column_name].median(), color='green', linestyle='dashed', label=f'Median: {dataframe[column_name].median():.2f}')

    plt.title(f"{dataset_name} - {column_name} Distribution")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    # Normality Test (Shapiro-Wilk)
    stat, p = shapiro(dataframe[column_name].dropna())
    print(f"\n Shapiro-Wilk Test for {column_name} in {dataset_name}: p-value = {p:.5f}")
    if p > 0.05:
        print(" Data is normally distributed (Use MEAN for imputation).")
    else:
        print(" Data is NOT normally distributed (Use MEDIAN for imputation).")

# Apply to both datasets
visualize_rating_distribution(tunisia_df, 'Rating', 'Tunisia Dataset')
visualize_rating_distribution(global_df, 'Rating', 'Global Dataset')
```

**handling missing values efficiently**

```python
# Handling Missing Values for Tunisia & Global Datasets (Future-proofed)

import pandas as pd

# Fill missing Ratings with the median
tunisia_df.loc[:, 'Rating'] = tunisia_df['Rating'].fillna(tunisia_df['Rating'].median())
global_df.loc[:, 'Rating'] = global_df['Rating'].fillna(global_df['Rating'].median())

# Fill missing Likes & Comments with average engagement per sentiment category
tunisia_df.loc[:, 'Likes'] = tunisia_df.groupby('Sentiment')['Likes'].transform(lambda x: x.fillna(x.mean()))
tunisia_df.loc[:, 'Comments'] = tunisia_df.groupby('Sentiment')['Comments'].transform(lambda x: x.fillna(x.mean()))

global_df.loc[:, 'Likes'] = global_df.groupby('Sentiment')['Likes'].transform(lambda x: x.fillna(x.mean()))
global_df.loc[:, 'Comments'] = global_df.groupby('Sentiment')['Comments'].transform(lambda x: x.fillna(x.mean()))

# Final Check for Missing Values
print("\n Final Check for Missing Values:")
print("Tunisia Dataset:\n", tunisia_df.isnull().sum())
print("\nGlobal Dataset:\n", global_df.isnull().sum())
```

**Checking for Inconsistencies in Categorical Columns**

```python
import re

#  Checking for inconsistencies in categorical columns
def check_unique_values(df, column_name):
    print(f"\n Unique values in '{column_name}':")
    print(df[column_name].unique())

# Checking for unique values for key categorical columns
for col in ["Sentiment", "Industry", "Source"]:
    check_unique_values(tunisia_df, col)
    check_unique_values(global_df, col)
```

**Standardizing Text-Based Data**

```python
# Standardize text-based data
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^\w\s.,!?]', '', text)  # Remove special characters except .,!?
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Apply text cleaning to 'Post/Review Text'
tunisia_df["Post/Review Text"] = tunisia_df["Post/Review Text"].apply(clean_text)
global_df["Post/Review Text"] = global_df["Post/Review Text"].apply(clean_text)

print("\n Text cleaning completed!")
```

**Merging the datasets in one dataset : 70% Tunisian data and 30% global data**

```python
# Sampling 30% of the global dataset
global_sample = global_df.sample(frac=0.30, random_state=42)

# Sampling 70% of the Tunisian dataset
tunisia_sample = tunisia_df.sample(frac=0.70, random_state=42)

# Merging the two sampled datasets
consumer_insights_df = pd.concat([tunisia_sample, global_sample], ignore_index=True)

# Display the final Consumer_Insights dataset
print("\n Merged Consumer Insights Dataset (30% Global, 70% Tunisia):")
display(consumer_insights_df.head())
```
